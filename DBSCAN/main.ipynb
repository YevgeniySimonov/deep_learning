```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import numpy as np
```

```python
# prepare dataset
X, y = make_moons(
    n_samples=200,
    noise=0.05,
    random_state=0
)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.tight_layout()
plt.show()
```

```python
db = DBSCAN(
    eps=0.2,
    min_samples=5,
    metric='euclidean'
)
y_db = db.fit_predict(X)
```

```python
num_clusters = np.unique(y_db).shape[0]
colors = ['lightblue', 'red', 'green', 'purple']
markers = ['o', 's', 'x', '.']
plt.figure()
for cluster_no, color, marker in zip(range(num_clusters), colors, markers):
    mask = y_db == cluster_no
    plt.scatter(
        X[mask, 0],
        X[mask, 1],
        c=color,
        edgecolors='black',
        marker=marker,
        s=40,
        label=f'Cluster {cluster_no}'
    )
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.tight_layout()
plt.show()
```
